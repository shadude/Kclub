```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
```

```python
def gen_mod_mtx(A):
    k = np.array([np.sum(A,axis = 1)])
    m = np.sum(k)
    B = A - (1/(2*m))*k.T@k
    return B
```

```python
A = np.eye(5)
B = gen_mod_mtx(A)
B
```

```python
G = nx.karate_club_graph()
A = nx.to_numpy_array(G)
B = gen_mod_mtx(A)
B
```
