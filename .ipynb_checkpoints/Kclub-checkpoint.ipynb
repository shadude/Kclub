```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
```

```python
def gen_mod_mtx(A):
    k = np.array([np.sum(A,axis = 1)])
    m = np.sum(k)
    B = A - (1/(2*m))*k.T@k
    return B
```

```python
A = np.eye(5)
B = gen_mod_mtx(A)
B
```

```python
G = nx.karate_club_graph()
A = nx.to_numpy_array(G)
B = gen_mod_mtx(A)
B
```

```python
e,v = np.linalg.eig(B)
```

```python
s = v[e.argmax(axis=0)]
```

```python
nodes = list(G.nodes)
n1 = []
n2 = []
for i in range(s.shape[0]):
    if s[i]>0:
        n1.append(nodes[i])
    else:
        n2.append(nodes[i])
n1
```

```python
g1 = G.subgraph(n1)
g2 = G.subgraph(n2)
nx.draw(G)
```

```python
cmap = plt.get_cmap('viridis')
colors = [cmap(i) for i in np.linspace(0,1,2)]
pos = nx.spring_layout(G)
nx.draw_networkx_nodes(G,pos,nodelist = g1, node_color = colors[0])
nx.draw_networkx_nodes(G,pos,nodelist = g2, node_color = colors[1])
nx.draw_networkx_edges(G,pos)
```
